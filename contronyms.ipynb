```
translator = google_translator()  
translated_en_hi=[]
# Note: Passing the entire sentence list yields _weird_ formatting errors.
# The only method that seems to work is literal for-looping over the individual sentences
for sentence_i in tqdm_notebook(list_sentence_she):
  translated_en_hi_i = translator.translate(sentence_i, lang_src='en',lang_tgt='hi')  
  translated_en_hi.append(translated_en_hi_i)
  
translated_hi_en=[]
for sentence_i in tqdm_notebook(translated_en_hi):
  translated_hi_en_i = translator.translate(sentence_i, lang_src='hi',lang_tgt='en')  
  translated_hi_en.append(translated_hi_en_i)
```

```
df_she2he=pd.DataFrame(data=np.stack([list_sentence_she,translated_en_hi,translated_hi_en]).T,
                       columns=['Original','En->Hi','Hi->En'])
df_she2he.to_csv('df_backtranslate_En2Hi2En.csv',index=False)
```

```
translated_en_ur=[]
for sentence_i in tqdm_notebook(list_sentence_she):
  translated_en_ur_i = translator.translate(sentence_i, lang_src='en',lang_tgt='ur')  
  translated_en_ur.append(translated_en_ur_i)
########################################################
translated_ur_en=[]
for sentence_i in tqdm_notebook(translated_en_ur):
  translated_ur_en_i = translator.translate(sentence_i, lang_src='ur',lang_tgt='en')  
  translated_ur_en.append(translated_ur_en_i)
```

```
df_she2he=pd.DataFrame(data=np.stack([list_sentence_she,translated_en_ur,translated_ur_en]).T,
                       columns=['Original','En->ur','ur->En'])
df_she2he.to_csv('df_backtranslate_En2ur2En.csv',index=False)
```
